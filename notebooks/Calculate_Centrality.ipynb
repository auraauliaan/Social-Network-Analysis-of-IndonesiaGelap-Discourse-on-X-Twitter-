```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/centrality-real.csv")
df
```

```python
top_closeness = df.sort_values('closnesscentrality', ascending=False).head(5)
top_closeness_filtered = top_closeness[['Id', 'Label', 'closnesscentrality']]
top_closeness_filtered
```

```python
top_betweenness = df.sort_values('betweenesscentrality', ascending=False).head(5)
top_betweenness_filtered = top_betweenness[['Id', 'Label', 'betweenesscentrality']]
top_betweenness_filtered
```

```python
top_eigenvector = df.sort_values('eigencentrality', ascending=False).head(5)
top_eigen = top_eigenvector[['Id', 'Label', 'eigencentrality']]
top_eigen
```
